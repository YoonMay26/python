```python
import pandas as pd
df_coffee = pd.read_csv("C:/Users/ysj20/hellopython/data/coffee_sales.csv")
df_coffee
```

```python
df_cash_menu = df_coffee[['cash_type','coffee_name']]
df_cash_menu
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_coffee['cash_type'].value_counts()
```

```python
# 시각화를 위한 데이터 준비
coffee_counts = df_coffee.groupby(['cash_type', 'coffee_name']).size().reset_index(name='count')

# Seaborn을 사용한 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=coffee_counts, x='coffee_name', y='count', hue='cash_type')
plt.title('Cash Type vs Coffee Name Count')
plt.xlabel('Coffee Name')
plt.ylabel('Count')
plt.legend(title='Cash Type')
plt.show()
```
